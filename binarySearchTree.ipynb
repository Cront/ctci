### Binary tree 
- A tree w/ @ most 2 children
- Exactly 1 root
- Exactly 1 path b/w root and any node
- (1 video said this but not the other) values of nodes in left subtree lesser than values of nodes in right subtree
- Very efficient (best case: O(nlog(n), worst case: O(n))

NOTE: only root node considered binary tree AND no node tree considered empty tree + binary tree

## Node class

### Depth First Value Func (Non-recursive)
Depth First Value values: [a -> b -> d -> e -> c -> f]
    
    Tree structure:

          a
         / \
        b   c
       / \   \
      d   e   f
      
- Uses a stack (pop from top, append from top)

```python
def depthFirstValues(root):
    if root is None: return list()
    
    result = list()
    stack = [root]
    
    while len(stack) > 0:
        current = stack.pop()
        result.append(current.val)
        
        if current.right: stack.append(current.right)
        if current.left: stack.append(current.left)
            
    return result
```

```python
depthFirstValues(a)
```

### Depth First Value Func (Recursive)

```python
def depthFirstValuesRecursive(root):
    if root is None: return []
    
    leftValues = depthFirstValuesRecursive(root.left)
    rightValues = depthFirstValuesRecursive(root.right)
    
    return [root.val, *leftValues, *rightValues] # * operator flattens the list and removes nested list struct
```

```python
depthFirstValuesRecursive(a)
```

### Breadth First Values
Breadth-first values: [a -> b -> c -> d -> e -> f]

Tree structure:

          a
         / \
        b   c
       / \   \
      d   e   f
      
- Uses a queue (enter from back, leave from front)

```python
def breadthFirstValues(root):
    if root is None: return list()
    
    values = list()
    queue = [root]
    
    while len(queue) > 0:
        curr = queue.pop(0)
        values.append(curr.val)
        if curr.left is not None: queue.append(curr.left)
        if curr.right is not None: queue.append(curr.right)
            
    return values
```

```python
breadthFirstValues(a)
```
