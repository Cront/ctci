#### Priority Queue

If T{1..6} are the tasks at hand, and the numbers with them are the priority of the tasks, priority queue returns high priorities first (T5, T2, T1, T6, T3, and T4). 

T1 - 5

T2 - 4

T3 - 7

T4 - 9

T5 - 2

T6 - 6


#### Heap

Want to be able to add to the priority queue too. Efficient to add and take out from priority queue.

Heap is an effective and efficient implementation of a priority queue. It is a tree data structure

Runtime to add elemnts to queue O(log(n))

#### Max heap


```python
import heapq
```

```python
data = [10,20,43,1,2,65,17,44,2,3,1]
heapq.heapify(data)
```

```python
data
```

#### NOTE
Confusion at organization of heap because it follows this tructure

i (root node)

(2*i)+1 (child node 1)

(2*1)+2 (child node 2)

(i-1)/2 (parent node)



#### heappop() returns highest priority node 

```python
heapq.heappop(data) 
```

#### Still maintains proper tree structure

```python
data
```

#### heappush() adds node with priority in proper tree structure

```python
heapq.heappush(data, 2)
heapq.heappush(data, 14)
heapq.heappush(data, 74)
```

```python
data
```

#### No go to way to make max heap (default is min heap)

What you can do is negate the heap 

```python
data = [-x for x in data]
```

```python
heapq.heapify(data)
```

```python
data
```
