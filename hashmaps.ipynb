Benefits of Hashtables

1. Custom keys are easier to work
2. Hashmaps allow for search in O(1), whereas arrays/linkedl ists are O(n)  

Keys must be an immutable data type

```python
cityMap = {} # could also be initialized with cityMap = dict()

cities = ['Calgary', 'Vancouver', 'Toronto']
cityMap['Canada'] = []
cityMap['Canada'] += cities
```

Dictionary: all keys need to be initialized

DefaultDict: all keys are already initialized

```python
from collections import defaultdict

city_map = defaultdict(list)
city_map['Canada'] += cities
```

3 ways to retrieve data from dicts
1. keys(): returns all keys from a dictionary in form of list
2. values(): returns all values from a dictionary in form of list 
3. items(): returns list of all key, value pairs in form of tuple
